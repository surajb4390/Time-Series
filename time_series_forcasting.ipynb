# Converting weekly data to Quarterly

```python
df_q = df.resample(rule='Q').mean()
df_q.head()
```

# Making Data Stationary

```python
df_q['diff_7']=df_q['Gasoline_Production']-df_q['Gasoline_Production'].shift(7)
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adfuller_Test(Temp):
    result=adfuller(Temp)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("Data is stationary")
    else:
        print("This data is non-stationary ")
```

```python
adfuller_Test(df_q['diff_7'].dropna())
```

After 7 differencing we got data is stationary

```python
df_q['diff_7'].plot(figsize=(16,6),color='k',linewidth=3)
```

# Model Implementation

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model=ARIMA(train_data['Gasoline_Production'],order=(5,1,4))
model_fit=model.fit()
```

```python
pred_arima=model_fit.predict(start=pred_start_date,end=pred_end_date,dynamic=True)
residuals=test_data['Gasoline_Production']-pred_arima
```

```python
df_q['Predicted_ARIMA']=pred_arima
```

```python
df_q[['Gasoline_Production','Predicted_ARIMA']].plot(ylim=[0,13000],figsize=(15,6))
```

Here ARIMA Not gives us best forcasting

```python
## creating a SARIMA model
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model_SARIMA=SARIMAX(train_data['Gasoline_Production'],order=(5,1,4),seasonal_order=(0,1,0,7))
```

```python
model_SARIMA_fit=model_SARIMA.fit()
```

```python
pred_Sarima=model_SARIMA_fit.predict(start=pred_start_date,end=pred_end_date,dynamic=True)
residuals=test_data['Gasoline_Production']-pred_Sarima
```

```python
df_q['Predicted_SARIMA']=pred_Sarima
```

```python
df_q[['Gasoline_Production','Predicted_SARIMA']].plot(ylim=[5000,13000],figsize=(15,6))
```

Observation:

            Here we can see that our test data perfectly fit on actual data

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(test_data['Gasoline_Production'],pred_Sarima))
```

```python
model_SARIMA=SARIMAX(df_q['Gasoline_Production'],order=(5,1,4),seasonal_order=(0,1,0,7)).fit()
df_q['Predictions'] = model_SARIMA.predict(start=0,end=len(df_q))
```
